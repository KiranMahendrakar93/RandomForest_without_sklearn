*  <font color='blue'><b> Write code for generating samples</b></font>

```
def generating_samples(input_data, target_data):

    '''In this function, we will write code for generating 30 samples '''
    # you can use random.choice to generate random indices without replacement
    # Please have a look at this link https://docs.scipy.org/doc/numpy-1.16.0/reference/generated/numpy.random.choice.html for more details
    # Please follow above pseudo code for generating samples

    # return sampled_input_data , sampled_target_data,selected_rows,selected_columns
    #note please return as lists

    # selecting and replicating row indices
    selected_rows = np.sort(np.random.choice(len(input_data), size = int(len(input_data) * 0.6), replace=False))
    replicated_rows = np.sort(np.random.choice(selected_rows, size = len(input_data) - len(selected_rows), replace = False))

    # selecting column numbers
    selected_cols = np.sort(np.random.choice(input_data.shape[1], size = np.random.randint(3,11), replace = False))

    # select data
    sample_data = input_data[selected_rows[:,None], selected_cols]
    sample_target_data = target_data[selected_rows]

    # Replicate data
    replicated_sample_data = input_data[replicated_rows[:, None], selected_cols]
    replicated_target_data = target_data[replicated_rows]

    # stack select and replicate data
    final_sample_data = np.vstack((sample_data, replicated_sample_data))
    final_target_data = np.vstack((sample_target_data.reshape(-1,1), replicated_target_data.reshape(-1,1)))

    return final_sample_data , final_target_data, selected_rows, selected_cols
```

```
def generate_n_samples(input_data, target_data, num_base_models):
    '''This function creates samples for each base models'''
    list_input_data =[]
    list_output_data =[]
    list_selected_rows= []
    list_selected_columns=[]


    for i in range(num_base_models):
        sampled_input_data , sampled_target_data, selected_rows, selected_cols = generating_samples(input_data, target_data)
        list_input_data.append(sampled_input_data)
        list_output_data.append(sampled_target_data)
        list_selected_rows.append(selected_rows)
        list_selected_columns.append(selected_cols)

    return list_input_data, list_output_data, list_selected_rows, list_selected_columns
```

```
# Use generating_samples function to create 30 samples
# store these created samples in a list

list_input_data, list_output_data, list_selected_rows, list_selected_columns = generate_n_samples(x,y,30)
print('Input list           : ', len(list_input_data), 'samples and each sample contain', len(list_input_data[0]), 'datapoints')
print('Output list          : ', len(list_output_data), 'samples and each sample contain', len(list_output_data[0]), 'targets')
print('Selected rows list   : ', len(list_selected_rows), 'samples and each sample contain', len(list_selected_rows[0]), 'unique datapoints')
print('Selected columns list: ', len(list_selected_columns), 'samples and each sample contain unique set of features')
```

*  <font color='blue'><b> Write code for building regression trees</b></font>

```
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

def build_base_model_trees(list_inputs, list_outputs, num_base_models):
    '''This function returns the list of base models'''

    list_all_models = []
    for i in range(num_base_models):
        model = DecisionTreeRegressor()
        list_all_models.append(model.fit(list_inputs[i], list_outputs[i]))

    return list_all_models
```

```
list_all_models = build_base_model_trees(list_input_data, list_output_data, 30)
print('Number of base models: ', len(list_all_models))
print('Number of features in each base model:')

num_features = []
for i, learner in enumerate(list_all_models):
    num_features.append(learner.n_features_)

print(num_features)
```

After getting predicted_y for each data point, we can use sklearns mean_squared_error to calculate the MSE between predicted_y and actual_y.

*  <font color='blue'><b> Write code for calculating MSE</b></font>

```
def mse_score(input_data, target_data, list_models, list_selected_columns):
    ''' This fucntion returns the mean squared error of true y and y predicted from base models'''

    models_y_pred = [] # To store predicted values of each model
    for i in range(len(list_models)):
        models_y_pred.append(list_models[i].predict(input_data[:, list_selected_columns[i]])) # Predict on each data point of corpus

    y_pred_median = np.median(models_y_pred, axis= 0)  # Median of y predicted using all models

    return mean_squared_error(target_data, y_pred_median)
```

```
mse_score(x,y, list_all_models, list_selected_columns)
```

<font color='blue'><b>Step - 3 </b></font>

Now calculate the $OOB Score =  \frac{1}{506}\sum_{i=1}^{506}(y^{i} - y^{i}_{pred})^{2}$.

*  <font color='blue'><b> Write code for calculating OOB score </b></font>

```
def oob_score(input_data, target_data, list_models, list_selected_rows, list_selected_columns):
    ''' This function returns the mean squared error of true y and y predicted using only base models with out-of-bag datapoints'''

    y_pred_median = [] # to store median of predicted y using oob models only
    for i, point in enumerate(input_data): # for each datapoint
        models_y_pred = []
        for j, model in enumerate(list_models):# for each base models
            if i not in list_selected_rows[j]: # predict base models with out-of-bag datapoints
                models_y_pred.append(model.predict(point[list_selected_columns[j]].reshape(1,-1)))

        y_pred_median.append(np.median(models_y_pred)) # Taking median of k base models

    return mean_squared_error(target_data, y_pred_median)
```

```
oob_score(x,y, list_all_models, list_selected_rows, list_selected_columns)
```

# <font color='blue'><b>Task 2</b></font>

### Computing CI of OOB Score and Train MSE

```
mse_scores = []
oob_scores = []
num_base_models = 30
for i in range(35):
    list_input_data, list_output_data, list_selected_rows, list_selected_columns = generate_n_samples(x, y, num_base_models)
    list_all_models = build_base_model_trees(list_input_data, list_output_data, num_base_models)
    mse_scores.append(mse_score(x, y, list_all_models, list_selected_columns))
    oob_scores.append(oob_score(x, y, list_all_models, list_selected_rows, list_selected_columns))

print(mse_scores)
print(oob_scores)
```

```
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1,2, figsize = (15,5))
sns.histplot(mse_scores, kde = True, ax= axes[0]).set_title('PDF of MSE errors')
sns.histplot(oob_scores, kde = True, ax=axes[1]).set_title('PDF of OOB errors')
plt.show()
```

```
# Non-parametric method: Bootstrapping
print('95% confidence interval of MSE_scores using non-parametric method: ', np.round(np.percentile(mse_scores, 2.5),3), np.round(np.percentile(mse_scores, 97.5),3))
print('95% confidence interval of OOB_scores using non-parametric method: ', np.round(np.percentile(oob_scores, 2.5),3), np.round(np.percentile(oob_scores, 97.5),3))
```
